```python
import pandas as pd
import matplotlib.pyplot as plt

# Define states and corresponding ages
states = [
    "Connecticut", "Georgia", "Maine", "Maryland", "Massachusetts",
    "New York", "New Hampshire", "New Jersey", "North Carolina" "Pennsylvania",
    "Vermont", "Tennessee", "Virginia", "West Virginia"
]
ages = [10, 25, 35, 40, 20, 45, 55, 65, 75, 85, 95, 50, 70]

# Create a DataFrame
df = pd.DataFrame({
    "State": states,
    "Missing Persons Age Group": ages
})

# Define age ranges
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '21-40', '41-60', '61-80', '81-100']

# Categorize states by age group
df['Age Range'] = pd.cut(df['Missing Persons Age Group'], bins=bins, labels=labels)

# Plot bar chart for states and ages
plt.bar(df['State'], df['Missing Persons Age Group'], color='orange', alpha=0.7)
plt.title('Missing Persons Age Group by State')
plt.xlabel('State')
plt.ylabel('Missing Persons Age Group')
plt.xticks(rotation=45, ha='right')
plt.show()

# Plot bar chart for age ranges
age_range_counts = df['Age Range'].value_counts(sort=False)
age_range_counts.plot(kind='bar', color='blue', alpha=0.7, figsize=(10, 6))
plt.title('Number of States by Age Range')
plt.xlabel('Age Range')
plt.ylabel('Number of States')
plt.xticks(rotation=45)
plt.show()
```
